```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
N = 50
bee_loc = np.array([25.0,25.0])
bee_vel = np.array([5,2])

bee_locs = []
images = []
for it in range(40):
    bee_vel = np.array([np.sin(it/5),np.cos(it/5)])*2.0
    bee_loc += bee_vel
    bee_locs.append(bee_loc.copy())
    image = 0.8+np.random.randn(N,N)*0.1
    image[int(bee_loc[0]),int(bee_loc[1])]+=0.4
    image[int(bee_loc[0])+1,int(bee_loc[1])]+=0.2
    image[int(bee_loc[0])-1,int(bee_loc[1])]+=0.2
    image[int(bee_loc[0]),int(bee_loc[1])+1]+=0.2
#    image[int(bee_loc[0]),int(bee_loc[1])-1]+=0.2
#    image[int(bee_loc[0])+1,int(bee_loc[1])+1]+=0.1
#    image[int(bee_loc[0])-1,int(bee_loc[1])-1]+=0.1
#    image[int(bee_loc[0])+1,int(bee_loc[1])-1]+=0.1
#    image[int(bee_loc[0])-1,int(bee_loc[1])+1]+=0.1
    #image[image>2.0] = 2.0
    image[image<0] = 0
    images.append(image)
```

```python
for i,image in enumerate(images):
    plt.figure()
    plt.matshow(image)
```

```python
partN = 100000
plt.figure(figsize=[9,9*10])
particles = np.random.rand(partN,4)
particles[:,0:2]*=N
particles[:,2:4] = np.random.randn(partN,2)
it = 0
for i,image in enumerate(images):
    bee_loc = bee_locs[i]
    particles[:,0:2] += particles[:,2:4]+np.random.randn(partN,2)*1.0 #x ~ p(x|x_old)    
    particles[:,2:4] += np.random.randn(partN,2)*0.5
    
    #TODO: sort out edges.
    particles[particles[:,0]>N-1,0] = N-1
    particles[particles[:,1]>N-1,1] = N-1
    particles[particles[:,0]<0,0]= 0
    particles[particles[:,1]<0,1] = 0

    w = np.array([image[int(p[0]),int(p[1])] for p in particles])-np.mean(image)
    w[w<0] = 0
    w = w**2
    w = w / np.sum(w)
    
    particles = particles[np.random.choice(len(w), partN, p=w)]
    
    
    if (i%4==0):
        it+=1
        plt.subplot(10,1,it)
        plt.plot(particles[0::15,0],particles[0::15,1],'.',markersize=1,alpha=0.01)
        plt.xlim([0,N])
        plt.ylim([0,N])
        plt.plot(bee_loc[0],bee_loc[1],'xk',markersize=10)
        #get pixel with most particles in.
        #find their mean velocity vector?
        plt.plot(np.mean(particles[:,0],0),np.mean(particles[:,1],0),'.g')
        plt.plot(np.mean(np.c_[particles[:,0],particles[:,0]+particles[:,2]],0),np.mean(np.c_[particles[:,1],particles[:,1]+particles[:,3]],0),'-g')
```
